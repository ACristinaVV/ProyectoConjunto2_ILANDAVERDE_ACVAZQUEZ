5. Use filter to determine the percentage of Fahrenheit temperatures in a list are within the range 32 to 80

### Respuesta Ivette

```python
np.random.seed(55555)
temperatures = np.random.uniform(25,110,30)

temperatura = list(filter(lambda x: 32 <= x <= 80, temperatures))

#ahora para saber el porcentaje de los que están en ese rango:
percentage = len(temperatura)/len(temperatures)
print("El porcentaje de temperaturas entre 32 y 80 Fahrenheit es:", percentage*100,"%")
```
